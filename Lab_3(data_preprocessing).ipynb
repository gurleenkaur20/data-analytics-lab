```python
df_car = pd.read_csv('car.csv')
df_car_copy = df_car.copy()
df_car.head()
```

```python
df_car.columns
```

```python
# changing the name of some columns
df_car = df_car.rename(columns={"Engine Fuel Type":"FuelType", "Engine HP":"HP", "Engine Cylinders":"Cylinders", "Transmission Type":"Transmission"})
```

```python
 columns_to_be_dropped = ['FuelType', 'Market Category', 'Vehicle Size', 'Vehicle Style']
```

```python
df_car.drop(columns_to_be_dropped, axis=1,  inplace=True)
```

```python
df_car.head()
```

```python
#checking duplicate values in the dataset
df_duplicated_records = df_car[df_car.duplicated()]
```

```python
df_duplicated_records.shape
```

```python
df_car.drop_duplicates(inplace=True)
```

```python
df_car.shape
```

```python
df_car.isnull().sum()
```

```python
df_car.dropna(inplace=True) #dropiing the records with missing values
```
